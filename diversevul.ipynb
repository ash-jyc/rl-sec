```python
# read diversevul_full.json, read fields, if cwe/cve not empty, write to diversevul_cwe_cve.json
# it's formatted like a dict in each line, but not a valid json list, so we need to read line by line
# write to valid json list format in output file
# check python in project or python in func

import json

input_file = 'diversevul_full.json'
output_file = 'diversevul_cwe_python.json'
output_data = []
with open(input_file, 'r') as f:
    for line in f:
        record = json.loads(line)
        if record.get('cwe') and ('python' in record.get('project') or 'python' in record.get('func')):
            output_data.append(record)
with open(output_file, 'w') as f:
    json.dump(output_data, f, indent=4)
```
